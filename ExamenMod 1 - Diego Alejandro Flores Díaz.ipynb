6. Una compa√±√≠a fabrica y venden dos modelos de l√°mpara L1 y L2. Para su fabricaci√≥n se necesita un trabajo manual de 20 minutos para el modelo L1 y de 30 minutos para el L2; y un trabajo de m√°quina para L1 y de 10 minutos para L2. Se dispone para el trabajo manual de 100 horas al mes y para la m√°quina 80 horas al mes. Sabiendo que el beneficio por unidad es de 15 y 10 euros para L1 y L2, respectivamente, planificar la producci√≥n para obtener el m√°ximo beneficio.

$$x = L1$$
$$y = L2$$

Nuestra funci√≥n objetivo es:
$$ f(x,y) = 15x + 10y $$

Las restricciones que tenemos son las siguientes:
$$1/3x + 1/2y \leq 100$$
$$1/3x + 1/6y \leq 80$$
$$x \geq 0$$
$$y \geq 0$$

```python
var ('x y')
```

```python
pulp.LpVariable
```

```python
mi_lp_problem = pulp.LpProblem ("Mi LP Ejercicio 6 Examen", pulp.LpMaximize)

x = pulp.LpVariable('x', lowBound=0, upBound= None, cat='Continuous')
y = pulp.LpVariable('y', lowBound=0, upBound= None, cat='Continuous')

mi_lp_problem += 15*x + 10*y

mi_lp_problem += 1/3*x + 1/2*y <= 100
mi_lp_problem += 1/3*x + 1/6*y <= 80
```

```python
mi_lp_problem
```

```python
mi_lp_problem.solve()
pulp.LpStatus[mi_lp_problem.status]
```

```python
print('Los valores que maximizan nuestra producci√≥n son: ')
for variable in mi_lp_problem.variables():
    print ('{} = {}'.format(variable.name, variable.varValue))
```

En el mercado s√≥lo se encuentra dos clases de compuestos: el tipo X con una composici√≥n de una unidad de A y 5 de B, y el otro tipo, Y, con una composici√≥n de cinco unidades de A y una de B. El precio del tipo X es de 10 euros y del tipo Y es de 30 ‚Ç¨. ¬øQu√© cantidades se han de comprar de cada tipo para cubrir las necesidades con un coste m√≠nimo?

$$x = Compuesto tipo x$$
$$y = Compuesto tipo y$$

Nuestra funci√≥n objetivo es:
$$f(x,y) = 10x + 30y$$

$$
\begin{array}{ll}
\mbox{Las restricciones son:} &  x  + 5y \geq 15 \\
       & 5x  + y \geq 15 \\
       & x \geq 0 \\
       & y \geq 0
\end{array}
$$

```python
pulp.LpVariable
```

```python
mi_lp_problem8 = pulp.LpProblem ("Mi LP Ejercicio 8 Examen", pulp.LpMinimize)

x = pulp.LpVariable('x', lowBound=0, upBound=None, cat='Continuous')
y = pulp.LpVariable('y', lowBound=0, upBound=None, cat='Continuous')

#objective function
mi_lp_problem8 += 10*x + 30*y

#constraints
mi_lp_problem8 += x + 5*y >= 15
mi_lp_problem8 += 5*x + 1*y >= 15
mi_lp_problem8
```

```python
mi_lp_problem8.solve()
pulp.LpStatus[mi_lp_problem8.status]
```
